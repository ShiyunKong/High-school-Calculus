```python
# Relu
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(x, 0)

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Function')
plt.show()
```

```python
#Sigmoid
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 1000)
y = sigmoid(x)

plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.show()
```

```python
# sin^4(x)
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return (np.sin(x))**4

# Generate x values
x = np.linspace(-1, 4, 1000)

# Generate y values
y = f(x)

# Plot the function
plt.plot(x, y)

# Set axis labels and title
plt.xlabel('x')
plt.ylabel('sin^4(x)')
plt.title('sin^4(x)')

# Show the plot
plt.show()
```

```python
#d'sigmoid
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

x = np.linspace(-10, 10, 1000)
y = sigmoid_derivative(x)

plt.plot(x, y)
plt.title('Derivative of Sigmoid Function')
plt.xlabel('x')
plt.ylabel("sigma'")
plt.show()
```

```python
#d'Custom
import numpy as np
import matplotlib.pyplot as plt

def d_sin4(x):
    return 4 * np.sin(x) ** 3 * np.cos(x)

x = np.linspace(-np.pi, np.pi, 1000)
y = d_sin4(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Derivative of sin^4(x)")
plt.show()
```

```python
#Integrate ReLU
import numpy as np
import matplotlib.pyplot as plt

# Define the function y = x
def f(x):
    return x

# Define the integration limits
a = 0
b = 2

# Evaluate the integral
integral = np.trapz(f(np.linspace(a, b)), np.linspace(a, b))

# Plot the function and the integrated region
x = np.linspace(0, 2)
y = f(x)
plt.plot(x, y, label='y = x')
plt.fill_between(x, y, 0, where=(x >= a) & (x <= b), alpha=0.5)
plt.text(1, 0.5, f"Integral = {integral:.2f}", fontsize=12)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#Integrate Sigmoid
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define integration limits
a, b = -2, 2

# Calculate the area under the curve using the quad() function
result, _ = quad(sigmoid, a, b)

# Generate x values for plotting
x = np.linspace(-5, 5, 100)

# Plot sigmoid function
plt.plot(x, sigmoid(x), label='Sigmoid')

# Fill area under the curve
x_fill = np.linspace(a, b, 100)
y_fill = sigmoid(x_fill)
plt.fill_between(x_fill, y_fill, 0, alpha=0.3)

# Add legend and title
plt.legend()
plt.title(f'Integrated region of sigmoid from ({a}, {b}), Area={result:.2f}')

# Show plot
plt.show()
```

```python
#Integrate Custom
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def integrand(x):
    return np.sin(x)**4

a, b = -2, 2  # Limits of integration

# Calculate the definite integral
area, _ = quad(integrand, a, b)

# Create x and y arrays for plotting the function
x = np.linspace(-2, 2, 1000)
y = integrand(x)

# Plot the function and the integrated region
fig, ax = plt.subplots()
ax.plot(x, y, label='Sin$^4$(x)')
ax.fill_between(x, y, where=((x>=a) & (x<=b)), alpha=0.3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Integrated Region of Sin$^4$(x)')
ax.legend()

# Display the calculated area on the plot
ax.text(1.5, 0.8, f'Area = {area:.4f}', fontsize=12)

plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load the dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Define the list of activation functions to compare
import numpy as np
def sin4(x):
    return tf.pow(tf.sin(x), 4)
activation_functions = ['relu', 'sigmoid', 'softmax',sin4]

# Define the model architecture
def build_model(activation_function):
    model = keras.Sequential([
    layers.Flatten(input_shape=(28,28), name='Input_layer'),
    layers.Dense(128, activation=activation_function, name='Hidden_layer_1'),
    layers.Dense(128, activation=activation_function, name='Hidden_layer_2'),
    layers.Dense(10, activation=tf.nn.softmax, name='Output_layer'),
])
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
# Train and evaluate the model with each activation function
for af in activation_functions:
    print(f'Training model with {af} activation function')
    model = build_model(af)
    model.summary()
    model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
    print(f'Test accuracy with {af} activation function: {test_acc}')
```
