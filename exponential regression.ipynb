```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/poly_fit.csv")
print(df)
```

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
```

```python
import numpy as np

```

```python
import scipy.optimize as opt
def func(t,a,b):
  return a*t**5+b*t**4
opt.curve_fit(func,  x,  y,  p0=(4, 0.1))
```

```python
import matplotlib.pyplot as plt
lst=[]
for i in x:
  print(i)
  y1=-1.29514705e-05*i**5 + 3.90695816e-04*i**4
  lst.append(y1)
plt.plot(x,lst,'b-' , label="polynomial regression")
plt.scatter(x,y,c='red', label='data')  
plt.legend(loc='best', labelspacing=0.1 )
plt.grid()
plt.title('Fit Ploynomial')
```
